```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
```

```python
pd.set_option('display.max_columns', None)
```

**Initial EDA - summary stats and plots of certain col distributions**

```python
# Load the dataset
df = pd.read_csv('../../data/transformed_data_men_21_to_24.csv')

# Display Summary Statistics
summary_statistics = df.describe()
display(summary_statistics)

# Plotting distributions for a selection of features
# Select a few columns to plot, given the large size of the dataset
columns_to_plot = ['team_score', 'opponent_score', 'luck', 'pace', 'choke_rate']

# Plot distributions
fig, ax = plt.subplots(len(columns_to_plot), 1, figsize=(10, 5 * len(columns_to_plot)))
for i, col in enumerate(columns_to_plot):
    sns.histplot(df[col], kde=True, ax=ax[i])
    ax[i].set_title(f'Distribution of {col}')
    ax[i].set_xlabel('')
    ax[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

**Initial EDA - data quality checks**

```python
# Check for missing values across all columns
missing_values = df.isnull().sum()

# Check for any duplicate rows in the dataset
duplicate_rows = df.duplicated().sum()

# Check for any negative values in 'team_score' or 'opponent_score' columns, which should not occur
negative_scores = df[(df['team_score'] < 0) | (df['opponent_score'] < 0)].shape[0]

# Displaying the results of these checks

print("Missing Values by Column:\n", sorted(dict(missing_values).items(), key=lambda x: x[1])[::-1])
print("\nNumber of Duplicate Rows:", duplicate_rows)
print("\nNumber of Negative Scores:", negative_scores)
```

**handling missing values**

```python
# Identifying numerical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Creating an imputer object with a strategy of replacing missing values with the median of the column
median_imputer = SimpleImputer(strategy='median')

# Applying the imputer to our numerical columns
df[numerical_cols] = median_imputer.fit_transform(df[numerical_cols])

# Checking if there are still missing values in these columns
missing_values_after_imputation = df[numerical_cols].isnull().sum()

print('missing_values_after_imputation:')
print(missing_values_after_imputation)

# Checking the number of unique values for each column that we imputed missing values for
unique_values_after_imputation = df[numerical_cols].nunique()

print('\nunique_values_after_imputation:')
print(unique_values_after_imputation)
```
